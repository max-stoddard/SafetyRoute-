```python
import iris
import os
 
# Connection parameters
username = 'demo'
password = 'demo'
hostname = os.getenv('IRIS_HOSTNAME', 'localhost')
port = '1972'
namespace = 'USER'
tableName = "SafeRoute.CrimeDataSample"
connection_string = f"{hostname}:{port}/{namespace}"# Connect to IRIS
conn = iris.connect(connection_string, username, password)
cursor = conn.cursor()
 
# SQL query template using standard SQL BETWEEN for filtering by bounding box.
sql = f""" 
    SELECT Longitude, Latitude, CrimeType, Context, ContextVector 
    FROM {tableName}
    WHERE Longitude BETWEEN ? AND ?
    AND Latitude BETWEEN ? AND ? 
    """

# Prepare a list to hold the output tuples: (list_of_crime_records, segment_length)
segment_results = []
 
# Iterate over each bounding box calculated with Shapely.
for (bbox, seg_length) in segment_bboxes:
    min_long, min_lat, max_long, max_lat = bbox
    
    # Execute the query by passing the bounding box limits.
    cursor.execute(sql, [min_long, max_long, min_lat, max_lat])
    crimes = cursor.fetchall()  # This returns a list of rows (tuples)
    segment_results.append((crimes, seg_length))
 
# Output the results for each segment.
for i, (crimes, seg_length) in enumerate(segment_results, start=1):
    print(f"Segment {i} (Length: {seg_length}): {len(crimes)} crimes found")
    for crime in crimes:
        print(crime)
    print("\n")
 
# Clean up: close the cursor and connection.
cursor.close()
conn.close()
```
